```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/devjaynemorais/modelos_descritivos_curso/main/Aula%2001/dados/players_match.csv')
```

```python
# Removendo linhas onde as coordenadas dos jogadores são NaN
df = df.dropna(subset=['Player_Position_X', 'Player_Position_Y'])
```

```python
# Preenchendo valores NaN das colunas center side
df['Center_Side_A_X'].fillna(25.0, inplace=True)
df['Center_Side_A_Y'].fillna(50.0, inplace=True)
```

```python
# Calculando distâncias para cada centro
df['Distancia_para_A'] = np.sqrt((df['Player_Position_X'] - df['Center_Side_A_X'])**2 + (df['Player_Position_Y'] - df['Center_Side_A_Y'])**2)
df['Distancia_para_B'] = np.sqrt((df['Player_Position_X'] - df['Center_Side_B_X'])**2 + (df['Player_Position_Y'] - df['Center_Side_B_Y'])**2)
```

```python
# Determinando o time com base na distância mínima
df['Time'] = np.where(df['Distancia_para_A'] < df['Distancia_para_B'], 'A', 'B')
```

```python
# Plotando as posições dos jogadores
plt.figure(figsize=(10, 7))
plt.scatter(df[df['Time'] == 'A']['Player_Position_X'], df[df['Time'] == 'A']['Player_Position_Y'], color='blue', label='Time A', s=100)
plt.scatter(df[df['Time'] == 'B']['Player_Position_X'], df[df['Time'] == 'B']['Player_Position_Y'], color='red', label='Time B', s=100)
plt.axvline(x=50, color='black', linestyle='--')  # Midfield line
plt.xlabel('Posição X')
plt.ylabel('Posição Y')
plt.title('Posição dos Jogadores no Campo')
plt.grid(True)
plt.show()
```
